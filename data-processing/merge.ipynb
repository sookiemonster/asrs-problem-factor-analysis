# Merge
A small notebook to merge all of the CSVs into a single, master CSV file

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import os
import glob
```

```python
directory_path = "../raw"

all_files = glob.glob(os.path.join(directory_path, "*.csv"))
li = []

# Read all CSVs 
for filename in all_files:
    df = pd.read_csv(filename, index_col=0, header=[0,1])
    li.append(df)

# Combine all DataFrames
df = pd.concat(li, axis=0)

# Remove 'unnamed' column at end (because the CSVs contain ',\n' by default)
df = df.iloc[:, :-1]
```

```python
df.to_csv(directory_path + "/merged_2022_2025.csv")
```
