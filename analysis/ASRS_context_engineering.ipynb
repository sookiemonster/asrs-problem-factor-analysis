```python
prompt_for_example = """
### INSTRUCTIONS:
1. Provide an "analysis" of the narratives by all problems or events that went wrong.
2. Referencing the CATEGORY DEFINITIONS, list the 3 most-likely categories for the primary problem.
3. Referencing your analysis, list of AT MOST 3 most-likely categories, and their respective AREAS FOR CONFUSION, provide an explanation for why the primary problem is the most logical/relevant choice.
4. Output the result in JSON.

You are completing the an incident entry with the following schema:
{{
  "report_number" : (the unique report number identifying the incident)
  "primary_problem": (the category you ARE ALREADY GIVEN),
  "analysis": "A brief 2-4 sentence step-by-step breakdown of problems or issues present in the narrative.",
  "most_likely_categories" : [A list of AT MOST 3 most-likely categories, including the category already given]
  "explanation": "3 sentences justifying why the primary problem is the most relevant given your analysis and why it is NOT one of the other most_likely_categories.",
}}

Under NO circumstance should you modify the primary problem or alter the report number.
ONLY update the "analysis", "most_likely_categories", and "explanation" properties.

For each incident, reference the narratives and category definitions to complete their corresponding answer:

"""

narrative_shard = """INCIDENT DATA:
Report Number: {acn}
Narratives:
{narratives}
"""

def make_example(narratives:str, acn:str | int, true_label:str ):
  """
  A way of getting examples + reasoning. We tell another model to generate the analysis and most_likely_categories based on the ground truth.
  """
  entry = {
      "report_number" : acn,
      "pred": true_label,
      "analysis": "",
      "most_likely_categories": "",
      "explanation": "",
  }
  return f"====== NEW INCIDENT ======\n{narrative_shard.format(acn=acn, narratives=narratives)}\nComplete the following: {json.dumps(entry)}\n====== END INCIDENT ======\n\n"
```

```python
def make_genai_example_prompt(batch_size=25):
  all_prompts = []
  header = simple_context + prompt_for_example
  footer = "NOW RETURN YOUR ANSWERS AS A LIST OF JSON ENTRIES ONLY IN A CODE BLOCK."

  prompt = header

  rows_added = 0
  for acn, row in nar.iterrows():

    label = ex_y.loc[acn]
    prompt += make_example(row['narratives'], acn, label)
    rows_added += 1

    if (rows_added % batch_size == 0 or rows_added == len(nar)):
      prompt += footer
      all_prompts.append(prompt)
      prompt = header

  return all_prompts

ex_prompts = make_genai_example_prompt()
print([x.count("NEW INCIDENT") for x in ex_prompts])
```

## Creating Few/Many Shot Prompt Context

```python
examples = pd.read_csv("labeled_examples.csv", index_col=0)
```

```python

def make_few_shot_example(narratives:str, acn:str | int, label:str, analysis:str, categories: str, explanation:str):
  """
  A way of getting examples + reasoning. We tell another model to generate the analysis and most_likely_categories based on the ground truth.
  """
  entry = {
      "report_number" : acn,
      "analysis": analysis,
      "most_likely_categories": categories,
      "explanation": explanation,
      "primary_problem": label,
  }
  return f"====== NEW INCIDENT ======\n{narrative_shard.format(acn=acn, narratives=narratives)}\nANSWER: {json.dumps(entry, indent=1)}\n====== END INCIDENT ======\n\n"

def format_few_shot_examples(examples:pd.DataFrame) -> str:
  res = ""

  for acn, row in examples.iterrows():
    res += make_few_shot_example(
        narratives=row['narratives'],
        acn=acn,
        analysis=row['analysis'],
        label=row['primary_problem'], categories=row['most_likely_categories'],
        explanation=row['explanation']
      )
  return res

print(format_few_shot_examples(examples))
```

```python
few_shot_simple_context = f"{simple_context}\nAs you classify each task, reference the following examples.\n\n### EXAMPLES\n{format_few_shot_examples(examples)}"

def make_few_shot_prompt(narratives:list[str], acn:str, context:str):
  return [
    {"role": "system", "content": context}, {"role": "user", "content": make_no_rule_user_prompt(narratives, acn)}
  ]

def make_few_shot_message_prompts(df:pd.DataFrame, context:str):
  res = df[df.columns].copy().apply(
      lambda x: make_few_shot_prompt(x.to_list(), acn=x.name, context=context),
      axis=1
  )
  return pd.DataFrame(res, columns=['prompt'])
```
